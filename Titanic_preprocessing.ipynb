```python
#importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
#shape of df
print(df.shape)
```

```python
#df head
print(df.head)
```

```python
#finding out how many are empty
print(df.isnull().sum())
```

```python
#cleaning the data

df.drop(columns=["Cabin"],inplace = True) #too much missing data so column dropped
df["Age"].fillna(df["Age"].median(), inplace = True) #filling missing ages with median age
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace = True) #filling missing embarked with mode
```

```python
df.isnull().sum()
#everything filled
```

```python
#creating a family size column

df["FamilySize"] = df["SibSp"] + df["Parch"] + 1 #plus 1 as when 0 parents and 0 siblings there is the person himself
df['FamilySize'].value_counts() #outputs number of people with the same family size
```

```python
#name column
print(df['Name'].head(10))
```

```python
#title extraction

df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
print(df['Title'].value_counts())
```

```python
#normalising titles
df['Title'].replace(["Mlle", 'Ms'], 'Miss', inplace = True) #they have same meanings
df['Title'].replace("Mme", "Mrs", inplace = True) #they have same meanings
df['Title'].replace(["Capt", "Col", "Don", "Dona", "Dr", "Jonkheer", "Lady", "Major", "Rev", "Sir", "Countess", "Master", "Countess"], "Other", inplace = True) #rare titles so a new field
print(df['Title'].value_counts())
```

```python
df.head()
```

```python
#Encoding
#no order of importance of the columns so imma use one hot encoding

df = pd.get_dummies(df, columns=['Sex','Title','Embarked'], drop_first= True, dtype=int) #dropfirst removes redundant columns
df.head()
```

```python
#scaling
print(df.describe())

from sklearn.preprocessing import StandardScaler, RobustScaler

scaler1 = StandardScaler()
scaler2 = RobustScaler()
df[['Age', "FamilySize"]] = scaler1.fit_transform(df[['Age', "FamilySize"]])
df[['Fare']] = scaler2.fit_transform(df[['Fare']]) #extreme outliers
df.head()
```

```python
df.to_csv('titanic_preprocessed', index = False)
```
